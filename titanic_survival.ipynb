```python
df=df.drop(columns=['PassengerId','Cabin','Name','Ticket'])
df
```

```python
df['Sex']=df['Sex'].replace({'male':1,'female':0})

df['Embarked']=df['Embarked'].replace({'S':1,'C':0,'Q':2})

df
```

```python
X=df.drop(columns=['Survived'])
y=df['Survived']
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=62)


model = RandomForestClassifier(n_estimators=100, random_state=62)


model.fit(X_train, y_train)
```

```python

y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)


print(f'Accuracy: {accuracy*100}%')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Classification Report:\n{classification_rep}')
```

```python
new_data = [[1, 0, 58.0, 0, 0, 26.5500, 1]]
new_predictions = model.predict(new_data)
if new_predictions== 1:
    print("The model predicts that the passenger survived.")
else:
    print("The model predicts that the passenger did not survive.")
print(new_predictions)
```
